```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# Поиск владельца по номеру документа
def people_document(num_doc):
    for document in documents:
        if document["number"] == num_doc:
            return f'Владелец документа: {document["name"]}'
            break
    else:
        return 'Документ не найден в базе'

# Поиск полки по номеру документа
def shelf_number(num_doc):
    if num_doc not in sum(directories.values(), []):
        return 'Документ не найден'
    else:
        for key, value in directories.items():
            if num_doc in value:
                return f'Документ расположен на полке {key}'

# Вывод полной информации
def list_documents():
    for document in documents:
        print(f'Тип документа: {document["type"]}, №: {document["number"]}, владелец: {document["name"]}, {shelf_number(document["number"])}')

# Добавить полку
def add_shelf_num(num_shelf):
    if num_shelf in directories.keys():
        return'Такая полка уже существует!'
    else:
        new_dict = {num_shelf: []}
        directories.update(new_dict)
        return 'Полка добавлена.'

# Удалить полку, если она пустая
def del_shelf_num(num_shelf):
    if num_shelf not in directories.keys():
        return 'Такой полки не существует!'
    elif num_shelf in directories.keys() and directories[num_shelf] == []:
        del(directories[num_shelf])
        return 'Полка удалена.'
    else:
        return 'На полке есть документы, удалите их перед удалением полки.'

# Вывести текущие полки
def print_shelfs():
    sh = ', '.join([i for i in directories.keys()])
    return f'Текущие полки {sh}'

# Главная функция
def main_menu():
    while True:
        command = input('Выберете одну из команд: p, s, l, as, ds. Для выхода наберите: q. \nВведите команду: ')
        if command == 'p':
            print(people_document(input('Узнать владельца по номеру документа: ')))
        elif command == 's':
            print(shelf_number(input('Узнать № полки по номеру документа: ')))
        elif command == 'l':
            print(list_documents())
        elif command == 'as':
            print(add_shelf_num(input('Укажите номер новой полки: ')))
            print(print_shelfs())
        elif command == 'ds':
            print(del_shelf_num(input('Укажите номер полки для удаления: ')))
            print(print_shelfs())
        elif command == 'q':
            print('Спасибо, что работу!')
            break
        else:
            print('Вы ввели команду не корректно, повторите ввод.')

print(main_menu())
```
